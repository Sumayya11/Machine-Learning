```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()

te_array = te.fit(data).transform(data)
te_array
```

```python
df = pd.DataFrame(te_array, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori, fpgrowth


purchase_item = "Milk"
ratio = 0.6

df1 = fpgrowth(df, min_support=ratio, use_colnames=True)
df1['itemsets'] = df1['itemsets'].apply(str)
df1 = df1[df1['itemsets'].str.contains(purchase_item)]

df2 = apriori(df, min_support=ratio, use_colnames=True, low_memory=True)
df2['itemsets'] = df2['itemsets'].apply(str)
df2 = df2[df2['itemsets'].str.contains(purchase_item)]

print("==============FpGrowth=============")
display(df1)

print("=============Aprori================")
display(df2)
```
